```python
from sympy import *
from ipywidgets import *
from math import factorial as f
from sympy.plotting import plot
import matplotlib.pyplot as plt
x=Symbol("x")
```

```python
def H(n):
    k=((-1)**n)*exp(x**2)
    d=diff(exp(-x**2),x,n)
    return(k*d)
```

```python
import numpy as np
```

```python
@interact
def psi_n(m=(0,0.1,0.001),w=(0,10),n=(0,7)):
    hbar=6.626*10**(-34)
    k=(m*w)
    y=k**(1/2)*x
    h=H(n).subs({x:y})/((2**n*f(n))**(1/2))
    u=(k/pi)**(1/4)*(exp(-y**2/2))
    psi=(h*u).evalf()
    
    b=np.linspace(-12,12,500)
    p=[]
    plt.figure(figsize=(7,5))
    for i in b:
        p.append(psi.subs({x:i}))
    V=m*(w**2)*(b**2)/2
    plt.plot(b,p,)
    plt.title("$\psi_{"+str(n)+"}(x)$")
    plt.grid(True)
    plt.xlabel("x")
    plt.xlim(-10,10)
    plt.axhline(color="k",alpha=0.5)
    plt.axvline(color="k",alpha=0.5)
    plt.legend(["m="+str(m)+"kg\n$\omega=$"+str(w)+"rad/s"+"\nk="+str(m*w**2)+"N/m"])
    plt.show()
    '''plt.figure(figsize=(7,5))

    plt.plot(b,np.array(p)**2)
    plt.title("$|\psi_{"+str(n)+"}(x)|^{2}$")
    plt.grid(True)
    plt.xlabel("x")
    plt.xlim(-9,9)
    plt.axhline(color="k",alpha=0.5)
    plt.axvline(color="k",alpha=0.5)
    plt.legend(["m="+str(m)+"kg\n$\omega=$"+str(w)+"rad/s"+"\nk="+str(m*w**2)+"N/m"])
    plt.show()
    plt.plot(b,V)
    plt.title("V(x)")
    plt.grid(True)
    plt.xlabel("x")
    plt.ylabel("$V(x)$")
    plt.axhline(color="k",alpha=0.5)
    plt.axvline(color="k",alpha=0.5)
    plt.legend(["m="+str(m)+"kg\n$\omega=$"+str(w)+"rad/s"+"\nk="+str(m*w**2)+"N/m"])
    plt.show()'''
```
